```python
def prepare_data_to_target_share_per_group(df: pd.DataFrame, target_column: str, evaluation_column: str) -> pd.DataFrame:
    """Function to aggregate data and calculate target share per group
    args:
        df: data frame with data
        tagret_column: target column stored in df
        evaluation_column: column to evaluate relation with target
    """
    if target_column not in df.columns:
        raise KeyError(f'There is no column called {target_column} in given data frame.')
    if evaluation_column not in df.columns:
        raise KeyError(f'There is no column called {evaluation_column} in given data frame.')
    aggregs = df[[target_column,evaluation_column]].groupby(evaluation_column).agg(['sum','count'])
    aggregs['high_income_share'] = aggregs.iloc[:,0] /aggregs.iloc[:,1]
    aggregs = aggregs.sort_values(by='high_income_share',ascending = False)
    return aggregs
```

```python
def plot_aggregation_target_share(aggregated_df: pd.DataFrame) -> None:
    """ Function to create bar plot with percentage 
        of positive examples in each group.
        args:
            aggregated_df: data frame which is an output 
            from prepare_data_to_target_share_per_group output 
    """
    if 'high_income_share' not in aggregated_df.columns:
        raise KeyError('The input df should be output from prepare_data_to_target_share_per_group function.')
    aggregated_df = aggregated_df.sort_values(by='high_income_share', ascending = False )
    fig = plt.figure(figsize = (10,6))
    plt.bar(x=aggregated_df.index, height=aggregated_df['high_income_share'])
    plt.title(f'Share of positive outcome in {aggregated_df.index.name}')
    
```

```python
# relacja zmiennych kategorycznych z targetem
aggregs = prepare_data_to_target_share_per_group(df=df,target_column='income_cat',evaluation_column='workclass')
plot_aggregation_target_share(aggregs)
```

### 3. Wybor metryki sukcesu i walidacji

Metryką sukcesu jest miara określająca jakość prognoz modelu. 
Metryk jest wiele, jednak w modelach klasyfikacyjnych może podzielić je na związane z:
- przypisaną przez model klasą
- przypisanym przez model prawdopobieństwem.

Skupimy się na pierwszym przypadku.
Do oceny modelu potrzebna będzie znajomość macierzy pomyłek (confusion matrix). Porównuje ona rzeczywiste klasy z tymi przypisanymi przez model.
<img src="figures/confusionmatrix.png">

$$Accuracy = \frac{TN+TP}{TN+TP+FN+FP}$$

Accuracy wyznacza odsetek dobrych dopasowań spośród wszystkich danych. Miara Accuracy może sprawdzić się dobrze przy zbilansowanych targecie (ok. 50% obserwacji w każdej z klas.)

Przeprowadźmy eksperyment. Wyliczymy accuracy modelu, który zawsze przypisuje klasę 0 (income <=50K).

```python
df['pred_0'] = 0
```

```python
from sklearn.metrics import accuracy_score, recall_score
```

```python
accuracy_score(df['income_cat'],df['pred_0'])
```

Wartości poniżej accuracy dla takiego modelu będą sugerować, że model nie działa, jednak może być to mylne i okazać się, iż faktycznie w porównywaniu obu klas nie radzi sobie najlepiej, ale może mieć wysoką jakoś w prognozowaniu klasy 1 (czego powyższy model nie robi).

Stąd wprowadzimy dodatkową miarę sukcesu modelu: TPR (true positive rate) (inaczej: sensitivity, recall)
$$ TPR = \frac{TP}{TP+FN} $$

<img src="figures/recall.jpeg">


TPR odpowiada na pytanie jaki odsetek z rzeczywiście pozytywnych (income >50K) udało się dobrze sklasyfikować.
Miar jest wiele i należy je dobierać według celu jaki chce się rozwiązać i na czym się skupić. 
Na potrzeby tych zajęć skupimy się na dwóch powyższych metrykach.

#### 4. Budowa modelu

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier, plot_tree
random_state=2021
```

```python
df.corr(method='spearman')['income_cat'].sort_values()
```

##### 4.1 Przygotowanie zmiennych do modelu.
Wstępne zmienne zostały przyporządkowane za pomocą analizy korelacji. Należy pamiętać, że jest jedynie wstępna selekcja jednowymiarowa, w przypadku niezadowalających rezulatatów, należy wrócić do selekcji.

```python
# przypisujemy X i y
feature_names = ['age','education-num','marital-status_Never-married','capital-gain','hours-per-week','relationship_Own-child',
'relationship_Not-in-family','relationship_Unmarried','relationship_Other-relative','occupation_Prof-specialty',
'occupation_Exec-managerial','occupation_Other-service','occupation_Handlers-cleaners']

X = df[feature_names]
y = df['income_cat']
```

##### 4.2 Podział zbioru na treningowy i testowy
Podział zbioru na treningowy i testowy jest bardzo ważnym aspektem. Nie możemy ufać jakości modelu jedynie na zbiorze treningowym (na tym, na którym się uczył), czyli na danych, które widział. Należy sprawdzić wyniki na zbiorze, którego model nie widział (zbiór testowy), to pozwala na ustalenie, czy model dobrze generalizuje i radzi sobie na nowych danych.

```python
# dzielimi na train i test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=random_state)

print(f"Train samples: {X_train.shape[0]}.")
print(f"Test samples: {X_test.shape[0]}.")
```

##### 4.3 Modelowanie
Modelowanie przeprowadzimy algorytmem drzewa decyzyjnego:
1. Drzewo decyzyjne

Drzewo decyzyjne, to algorytm, który dzieli nasze dane według zestawu reguł. Mają one na celu rodzielenie danych na podgrupy, które będą jak najbardziej jednolite z uwagi na target.
<center>
<img src="figures/definicja_drzew_wizualizacja.png">
</center>

```python
# definicja modelu
DT_model = DecisionTreeClassifier(random_state=random_state, max_depth=3, min_samples_leaf=20)
```

```python
# fitowanie modelu (dopasowanie do danych)
DT_model.fit(X_train,y_train)
```

```python
plot_tree(DT_model,feature_names= feature_names)
plt.show()
```

```python
# Zwiększymy maksymalną głębokość drzewa
DT_model2 = DecisionTreeClassifier(random_state=random_state, max_depth=10, min_samples_leaf=20)
DT_model2.fit(X_train,y_train)
```

##### 4.4 Ocena modeli
Porównamy oba modele z uwagi na accuracy. 
Najpierw jednak musimy wyznaczyć predykcje.

```python
DT_model2_pred_train = DT_model2.predict(X_train)
DT_model2_pred_test = DT_model2.predict(X_test)
```

```python
DT_model_pred_train = DT_model.predict(X_train)
DT_model_pred_test = DT_model.predict(X_test)
```

```python
accuracy_score(DT_model2_pred_train, y_train)
```

```python
accuracy_score(DT_model2_pred_test, y_test)
```

```python
accuracy_score(DT_model_pred_train, y_train)
```

```python
accuracy_score(DT_model_pred_test, y_test)
```

```python
confusion_matrix(DT_model2_pred_test, y_test)
```
